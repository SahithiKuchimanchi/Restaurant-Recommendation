```python
initial_rows = df.shape[0]
df.dropna(subset=['Cuisines'], inplace=True)
final_rows = df.shape[0]
print(f"Missing values in 'Cuisines' handled. Removed {initial_rows - final_rows} rows.")
print(f"New shape of DataFrame: {df.shape}")
```

```python
cols_to_drop = [col for col in df.columns if df[col].nunique() == 1]
if cols_to_drop:
    df.drop(columns=cols_to_drop, inplace=True)
    print(f"Removed columns with only one unique value: {cols_to_drop}")
else:
    print("No columns with only one unique value found.")
print(f"New shape of DataFrame: {df.shape}")
```

```python
binary_cols = ['Has Table booking', 'Has Online delivery', 'Is delivering now']
for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].map({'Yes': 1, 'No': 0})
        print(f"Column '{col}' converted to 1/0 numerical representation.")
    else:
        print(f"Column '{col}' not found in DataFrame.")

print("First 5 rows with updated binary columns:")
print(df[binary_cols].head())
```

```python
country_codes = {
    1: 'India',
    216: 'United States',
    215: 'United Kingdom',
    30: 'Brazil',
    189: 'South Africa',
    214: 'United Arab Emirates',
    148: 'New Zealand',
    208: 'Turkey',
    14: 'Australia',
    162: 'Philippines',
    94: 'Indonesia',
    166: 'Qatar',
    184: 'Singapore',
    191: 'Sri Lanka',
    37: 'Canada'
}
df['Country Name'] = df['Country Code'].map(country_codes)
print("New 'Country Name' column created by mapping 'Country Code'.")
print("First 5 rows with new 'Country Name' column:")
print(df[['Country Code', 'Country Name']].head())
```

```python
df['Cuisines'] = df['Cuisines'].apply(lambda x: [c.strip() for c in x.split(',')])
print("Cuisines column processed: comma-separated strings split into lists.")
print("First 5 rows of 'Cuisines' column after splitting:")
print(df['Cuisines'].head())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Instantiate MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Fit and transform the 'Cuisines' column
cuisine_features = mlb.fit_transform(df['Cuisines'])

print("Cuisine features vectorized using MultiLabelBinarizer.")
print(f"Shape of cuisine_features: {cuisine_features.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(cuisine_features)

print("Cosine similarity matrix calculated.")
print(f"Shape of cosine_sim: {cosine_sim.shape}")
print("First 5x5 rows and columns of cosine_sim:")
print(cosine_sim[:5, :5])
```

```python
indices = pd.Series(df.index, index=df['Restaurant Name']).drop_duplicates()

def get_recommendations(restaurant_name, cosine_sim=cosine_sim, df=df, indices=indices):
    # Get the index of the restaurant that matches the name
    if restaurant_name not in indices:
        return "Restaurant not found in the dataset."

    idx = indices[restaurant_name]

    # Get the pairwise similarity scores for all restaurants with that restaurant
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the restaurants based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar restaurants
    # Exclude the first element as it is the restaurant itself
    sim_scores = sim_scores[1:11]

    # Get the restaurant indices
    restaurant_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar restaurants with relevant details
    recommendations_df = df.iloc[restaurant_indices][['Restaurant Name', 'Aggregate rating', 'Cuisines']]
    return recommendations_df

print("Recommendation function 'get_recommendations' defined.")
```

```python
print("Generating recommendations for 'Le Petit Souffle':")
recommendations = get_recommendations('Le Petit Souffle')
print(recommendations)

print("\nGenerating recommendations for 'Ooma':")
recommendations = get_recommendations('Ooma')
print(recommendations)
```
